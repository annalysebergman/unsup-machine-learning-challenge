# Unsupervised Machine Learning Challenge

### Part 1: Prepare the data

```python
import pandas as pd
import pathlib as path
from sklearn.preprocessing import StandardScaler
import numpy as np
import sklearn.cluster as KMeans
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
# read in dataset into pandas dataframe
myopia_df = pd.read_csv("myopia.csv")
myopia_df
```

```python
# Remove the "MYOPIC" column from the dataset.
myopia_df = myopia_df.drop(columns = ['MYOPIC'])
myopia_df
```

```python
# standarize columns
scaler = StandardScaler()
myopia_scaled = scaler.fit_transform(myopia_df)
```

### Part 2: Apply Dimensionality Reduction

```python
# Perform dimensionality reduction with PCA. How did the number of the features change?
pca = PCA(n_components=.9)
myopia_pca = pca.fit_transform(myopia_scaled)
```

```python
# Further reduce the dataset dimensions with t-SNE and visually inspect the results. 
# To do this, run t-SNE on the principal components, which is the output of the PCA transformation.
tsne = TSNE(learning_rate=250)
myopia_tsne = tsne.fit_transform(myopia_pca)
myopia_tsne.shape
```

```python
# Create a scatter plot of the t-SNE output. Are there distinct clusters?
plt.scatter(myopia_tsne[:,0], myopia_tsne[:,1])
plt.show()
```
